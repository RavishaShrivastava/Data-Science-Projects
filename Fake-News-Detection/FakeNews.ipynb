```python
#importing necessary libraries
import pandas as pd
```

```python
true = pd.read_csv('True.csv')
```

```python
fake = pd.read_csv('Fake.csv')
```

```python
true.head()
```

```python
fake.head()
```

```python
true.info()
```

```python
fake.info()
```

```python
#labeling data
true['label'] = 1
fake['label'] = 0
```

```python
true.head()
```

```python
fake.head()
```

```python
#concatenation of both the datasets
data = pd.concat([true,fake],axis=0)
data.head()
```

```python
data.tail()
```

```python
#check null values
data.isnull().sum()
```

```python
#dropping columns
data = data.drop(['title','subject','date'],axis=1)
```

```python
#Needs shuffling of rows
data = data.sample(frac=1)
```

```python
data.head()
```

```python
data = data.reset_index(drop=True)
```

```python
data.head()
```

```python
#handling text column
import re
def handle_text(text):
  #To lower
  text = text.lower()
  #Removing digits
  text = re.sub(r'\d','',text)
  #remove URLs
  text = re.sub(r'http\S+','',text)
  #remove special characters
  text = re.sub(r'[^\w\s]','',text)
  #remove new line
  text = re.sub(r'\n','',text)
  return text
```

```python
data['text'] = data['text'].apply(handle_text)
```

```python
data['text']
```

```python
x = data['text'] #Dependent variable
y = data['label'] #Independent variable
```

```python
#handling categorical values
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
x_train = vectorization.fit_transform(x_train)
x_test = vectorization.transform(x_test)
```

Model 1 - Logistics Regression


```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
#evaluation of model
from sklearn.metrics import accuracy_score
print("Training accuracy: " ,accuracy_score(lr.predict(x_train),y_train))
print("Testing accuracy: " ,accuracy_score(lr.predict(x_test),y_test))
```

Model 2 - Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
print("Training accuracy: ", accuracy_score(y_train, dt.predict(x_train)))
print("Testing accuracy: ", accuracy_score(y_test, dt.predict(x_test)))
```

Model 3 - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
print("Training accuracy: ", accuracy_score(y_train,rf.predict(x_train)))
print("Testing accuracy: ", accuracy_score(y_test, rf.predict(x_test)))
```

Model 4 - Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier()
gb.fit(x_train,y_train)
```

```python
print("Training accuracy: ", accuracy_score(y_train,gb.predict(x_train)))
print("Testing accuracy: ", accuracy_score(y_test, gb.predict(x_test)))
```

```python
def output_label(n):
  if n==1:
    return "ITS A TRUE NEWS"
  elif n==0:
    return "ITS A FAKE NEWS"
  else:
    return "UNKNOWN"
```

```python
def manual_testing(news):
  testing_news = {"text":[news]}
  new_def_test = pd.DataFrame(testing_news)
  new_def_test["text"] = new_def_test["text"].apply(handle_text)
  new_x_test = new_def_test["text"]
  new_xv_test = vectorization.transform(new_x_test)
  predict_lr = lr.predict(new_xv_test)
  predict_gb = gb.predict(new_xv_test)
  predict_rf = rf.predict(new_xv_test)

  return "LR prediction: ", output_label(predict_lr[0]), "GB prediction: ", output_label(predict_gb[0]), "RF prediction: ", output_label(predict_rf[0])
```

```python
news = str(input())
```

```python
manual_testing(news)
```

```python
import joblib

joblib.dump(lr, 'logistic_model.pkl')
joblib.dump(gb, 'gradient_boosting_model.pkl')
joblib.dump(rf, 'random_forest_model.pkl')
joblib.dump(vectorization, 'vectorizer.pkl')
```
